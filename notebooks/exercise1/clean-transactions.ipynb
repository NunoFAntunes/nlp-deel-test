```python
import pandas as pd
import string
from unidecode import unidecode
import re
pd.set_option("display.max_colwidth", 10000)
```

```python
TRANSACTIONS_DATA_PATH = '../../data/transactions.csv'
df_transactions = pd.read_csv(TRANSACTIONS_DATA_PATH)
```

```python
def remove_punctuation(s):
    return s.translate(s.maketrans({k: ' ' for k in string.punctuation})).lower()
```

```python
# It deals with accents and names in different writing systems
def translate_unicode_into_ascii(s):
    return unidecode(s)
```

```python

df_transactions['description'] = df_transactions['description'].apply(translate_unicode_into_ascii)
df_transactions['description'] = df_transactions['description'].apply(lambda x:x.casefold())
```

```python
def remove_reference(s):
    r = "ref .{10,25}\/\/.{10,20}\/\/c\s*n\s*t\s*r|ref .{10,25}\/|\d{9,15}\/\/c\s*n\s*t\s*r"
    return re.sub(r, '', s)
```

```python
def remove_other_tokens(s):
    r = "for deel|to deel|deel|from|payment|transfer|received|\d{1,10}"
    return re.sub(r, '', s)
```

```python
def remove_extra_whitespaces(s):
    remove_end_spaces = "^\s+|\s+$"
    remove_double_spaces = "\s\s+"
    s = re.sub(remove_end_spaces, '', s)
    return re.sub(remove_double_spaces, ' ', s)
```

```python
df_transactions['description'] = df_transactions['description'].apply(remove_reference)
```

```python
df_transactions['description'] = df_transactions['description'].apply(remove_punctuation)
```

```python
print(df_transactions['description'].to_string(index=False))
```

```python
df_transactions['description'] = df_transactions['description'].apply(lambda x:re.sub(' +', ' ', x))
```

```python
df_transactions['description'] = df_transactions['description'].apply(remove_other_tokens)
```

```python
df_transactions['description'] = df_transactions['description'].apply(remove_extra_whitespaces)
```

```python
print(df_transactions['description'].to_string(index=False))
```

```python
df_transactions.to_csv('../../data/transactions_clean.csv', index=False)
```
